<b> Write code for generating samples</b><

```
def generating_samples(input_data, target_data):
    '''In this function, we will write code for generating 30 samples '''

    sel_rows = np.random.choice(range(506),303,replace = False)
    repl_rows = np.random.choice(range(303),203,replace = False)
    sel_cols = np.random.choice(range(13),10,replace = False)

    sample_data = input_data[sel_rows[:, None], sel_cols]
    target_sample = target_data[sel_rows]


    # Replicate Rows
    repl_sample_data = sample_data[repl_rows]
    repl_target_sample = target_sample[repl_rows]

    # Concat
    final_sample = np.vstack((sample_data,repl_sample_data))
    final_target_sample = np.vstack((target_sample.reshape(-1,1),repl_target_sample.reshape(-1,1)))
    

    return final_sample , final_target_sample,sel_rows,sel_cols
```

<font color='cyan'> <b> Grader function - 1 </b> </fongt>

```
def grader_samples(a,b,c,d):
    length = (len(a)==506  and len(b)==506)
    sampled = (len(a)-len(set([str(i) for i in a]))==203)
    rows_length = (len(c)==303)
    column_length= (len(d)>=3)
    assert(length and sampled and rows_length and column_length)
    return True
a,b,c,d = generating_samples(x, y)
grader_samples(a,b,c,d)
```

*  <font color='blue'> <b>Create 30 samples </b>

![alt text](https://i.imgur.com/p8eZaWL.jpg)

```
# Use generating_samples function to create 30 samples 
# store these created samples in a list
list_input_data =[]
list_output_data =[]
list_selected_row= []
list_selected_columns=[]

for i in range(0,30):
  a,b,c,d = generating_samples(x, y)
  list_input_data.append(a)
  list_output_data.append(b)
  list_selected_row.append(c)
  list_selected_columns.append(d)
```

<font color='orange'><b>Flowchart for calculating MSE </b></font>

![alt text](https://i.imgur.com/sPEE618.png)

After getting predicted_y for each data point, we can use sklearns mean_squared_error to calculate the MSE between predicted_y and actual_y.

*  <b> Write code for calculating MSE</b>

```
def cal_MSE(x,models,list_selected_row_,list_selected_columns_):
  y_preds = []
  for x_pt in x:
    temp_y = []
    for i in range(30):
      # data_point = x[i]
      sel_cols = list_selected_columns_[i]
      res = models[i].predict(x_pt[sel_cols].reshape(1,-1))
      temp_y.append(res)

    y_preds.append(np.median(temp_y))
  
  return y_preds
```

```
y_preds = cal_MSE(x,model_list,list_selected_row,list_selected_columns)
MSE = mean_squared_error(y,y_preds)
print("MSE : ",MSE)
```

<font color='blue'><b>Step - 3 </b></font>

<font color='orange'><b>Flowchart for calculating OOB score</b></font>

Now calculate the $OOB Score =  \frac{1}{506}\sum_{i=1}^{506}(y^{i} - y^{i}_{pred})^{2}$.

*  <b> Write code for calculating OOB score </b>

```
def cal_OOB(x,models,list_selected_row_,list_selected_columns_):
  y_preds_oob = []
  for x_idx,x_pt in enumerate(x):
    temp_y = []
    for i in range(30):
      sel_rows = list_selected_row_[i]
      # if data_point not in dataset : pass 
      if x_idx in sel_rows:
        continue

      sel_cols = list_selected_columns_[i]
      res = models[i].predict(x_pt[sel_cols].reshape(1,-1))
      temp_y.append(res)

    y_preds_oob.append(np.median(temp_y))

  return y_preds_oob
```

```
y_preds_oob = cal_OOB(x,model_list,list_selected_row,list_selected_columns)
OOB = mean_squared_error(y,y_preds_oob)
print("OOB Score : ",OOB)
```

# <font color='blue'><b>Task 2</b></font>

```
from tqdm import tqdm
```

```
def run_task_1(n):
  mse_list = []
  oob_list = []
  for i in tqdm(range(n)):
    # Use generating_samples function to create 30 samples 
    list_input_data_ =[]
    list_output_data_ =[]
    list_selected_row_= []
    list_selected_columns_=[]
    for i in range(0,30):
      a_,b_,c_,d_ = generating_samples(x, y)
      list_input_data_.append(a_)
      list_output_data_.append(b_)
      list_selected_row_.append(c_)
      list_selected_columns_.append(d_)


    #Building Decision Trees
    model_list_new = []
    for idx in range(30):
      model = DecisionTreeRegressor(max_depth = None)
      model.fit(list_input_data_[idx], list_output_data_[idx])
      model_list_new.append(model)

    #MSE & OOB
    mse = mean_squared_error(y,cal_MSE(x,model_list_new,list_selected_row_,list_selected_columns_))
    oob = mean_squared_error(y,cal_OOB(x,model_list_new,list_selected_row_,list_selected_columns_))

    mse_list.append(mse)
    oob_list.append(oob)

  return mse_list,oob_list
  
```

```
mse_list,oob_list = run_task_1(30)
```

${\text{Upper 95%  limit}{\displaystyle ={\bar {x}}+{\text{SE}}\times 1.96}}$ <br>
${\text{Lower 95%  limit}{\displaystyle ={\bar {x}}-{\text{SE}}\times 1.96}}$


```
# Z Table using scipy : https://stackoverflow.com/questions/20864847/probability-to-z-score-and-vice-versa
def cal_CI(sample,ci_percent):
  x_mean = np.mean(sample)
  x_std = np.std(sample)
  SE = x_std/np.sqrt(len(sample))

  z_score = st.norm.ppf(ci_percent/100)

  upper_lim = x_mean + (z_score*SE)
  lower_lim = x_mean - (z_score*SE)

  return lower_lim,upper_lim
```

```
from prettytable import PrettyTable
table = PrettyTable()
table.field_names = ["METRIC", "CONFIDENCE PERCENTAGE", "LOWER","VALUE"]

hp = [98,97,96,95,90,80,70]
for p in hp:
  low, up = cal_CI(mse_list,p)
  table.add_row(["MSE", str(p)+"%", np.round(low,6), np.round(up,6)])
for p in hp:
  low, up = cal_CI(oob_list,p)
  table.add_row(["OOB", str(p)+"%", np.round(low,6), np.round(up,6)])

print(table)
```
