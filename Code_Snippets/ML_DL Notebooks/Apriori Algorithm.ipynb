```python
import numpy as np
import matplotlib.pyplot as mtp  
import pandas as pd  


import apyori 
```

### Read Data

```python
store_data = pd.read_csv("store_data.csv", header=None)
display(store_data.head())
print(store_data.shape)
```

```python
records = []
for row in store_data.itertuples():
  records.append(np.array([str(x).strip() for x in list(row[1:]) if x not in [np.NaN]]))
```

```python
for x in records[56:60]:
  print(x)
```

## Apriori Algorithm

```python
association_rules = apyori.apriori(records, min_support=0.0035, min_confidence=0.1, min_lift=2, min_length=2)
association_results = list(association_rules)
```

```python
print("There are {} Relation derived.".format(len(association_results)))
```

### Frequent Itemsets

```python
print("DERIVED FREQUENT ITEMSETS")
for i in range(0, 20):
    print(association_results[i][0])
```

### Derived Association Rules

```python
association_dict = {"Rule":[],"Antecedant":[],"Precedent":[],"Support":[],"Confidence":[],"Lift":[]}

for item in association_results:
    # first index of the inner list
    # Contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    # print("Rule: " + items[0] + " -> " + items[1])
    association_dict["Rule"].append(f"{items[0]} => {items[1]}")
    association_dict["Antecedant"].append(items[0])
    association_dict["Precedent"].append(items[1])
    

    # second index of the inner list
    # print("Support: " + str(item[1]))
    association_dict["Support"].append(str(item[1]))

    # third index of the list located at 0th
    # of the third index of the inner list

    # print("Confidence: " + str(item[2][0][2]))
    # association_results["Confidence"].append(str(item[2][0][2]))
    # print("Lift: " + str(item[2][0][3]))
    association_dict["Lift"].append(str(item[2][0][2]))
    association_dict["Confidence"].append(str(item[2][0][3]))
    # print("=====================================")
  
assoc_df = pd.DataFrame(association_dict)
assoc_df.head()
```

```python
print("Number of rules created for each Product Bought")
assoc_df.groupby("Antecedant")["Rule"].count().sort_values()
```

```python
# possible prediction
def predict_possible(purchased_product):
  return assoc_df[assoc_df["Antecedant"] == purchased_product].sort_values(["Confidence"])


res_df = predict_possible("frozen vegetables")
res_df
```
